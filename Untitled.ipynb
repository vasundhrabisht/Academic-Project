## Detect outliers

```python
df2=pd.read_excel("Vasundi book 1.xlsx")
```

```python
df2.drop(["Corrected sales"],axis=1,inplace=True)
```

```python
df2 = df2.rename(columns={"date": "time", "Corrected sales": "value"})
ts1 = TimeSeriesData(df2)
```

```python
from kats.detectors.outlier import OutlierDetector

outlier_detector = OutlierDetector(ts1, "multiplicative")

outlier_detector.detector()
outliers = outlier_detector.outliers
outliers[0]
```

## Correcting outliers by interpolation

```python
ts_day_outliers_interpolated = outlier_detector.remover(interpolate=True)
```
